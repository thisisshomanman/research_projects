```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

# 
undrunk_path = '/Users/shomak/Documents/labo/data_for_histgram/undrunk.png'
drunk_path = '/Users/shomak/Documents/labo/data_for_histgram/drunk.png'

undrunk_img = Image.open(undrunk_path).convert('L')  # 'L'
drunk_img = Image.open(drunk_path).convert('L')

# numpy
undrunk_data = np.array(undrunk_img)
drunk_data = np.array(drunk_img)

# 
undrunk_hist, _ = np.histogram(undrunk_data.flatten(), bins=256, range=[0,256])
drunk_hist, _ = np.histogram(drunk_data.flatten(), bins=256, range=[0,256])

# 
plt.figure(figsize=(10, 6))
plt.title('Histogram Comparison', y=-0.18)
plt.xlabel('Pixel Value (Blackness)', loc='right')
plt.ylabel('Frequency',  loc='top', rotation=360)
plt.plot(undrunk_hist, color='blue', label='Undrunk Picture')
plt.plot(drunk_hist, color='red', label='Drunk Picture')
plt.legend(loc="upper right")
plt.tight_layout()
plt.show()
```
